```python
import requests
# You need to install the requests module to use this code
import json   

api_key = False
# If you have a Google Places API key, enter it here
# api_key = 'AIzaSy___IDByT70'
# https://developers.google.com/maps/documentation/geocoding/intro

#API
if api_key is False:
    api_key = 42
    serviceurl = 'http://py4e-data.dr-chuck.net/json'
else :
    serviceurl = 'https://maps.googleapis.com/maps/api/geocode/json'

#File handling
in_file = open("Prac.csv","r")

lines = in_file.readlines()
d = {}

out_file = open("OutputPrac.csv","w")
out_file.write('Address, Latitude, Longitude')
out_file.write('\n')

# loop accesses each element of list, gets content, parsing in json and responding with that value corresponding 
# key of the dictionary as needed 
for line in lines[1:]:
    line = line.strip()
    address = line
    payload = dict()
    payload['address'] = address
    if api_key is not False: payload['key'] = api_key #needed for the original API
 
    r = requests.get(serviceurl, params=payload)
    data = r.text
    
    try:
        js = json.loads(data)
    except:
        js = None
        
    if line not in d:
        try:
            lat = js['results'][0]['geometry']['location']['lat']
            lng = js['results'][0]['geometry']['location']['lng']
            d[line] = [lat, lng]
            str = '{},{},{}'.format(line, d[line][0], d[line][1])
            if 20.86382<lat<26.33338 and 88.15638<lng<92.30153:
                print(str)
            else:
                print(address + 'WRONG DATA') #data discrepancy
        except:
            print(address + 'NOT FOUND') #data not in API db
            
        out_file.write(str)
        out_file.write('\n')
       

in_file.close()
out_file.close()

print("FINISHED")
```
